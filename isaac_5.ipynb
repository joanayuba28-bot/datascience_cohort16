# TASK 1: Data Collection

```python
# user input.
cities = []
for i in range(5):
    city_name = input(f"Enter the name of city {i + 1}: ")
    cities.append(city_name)
print(f"You selected these cities: {cities}")
```

# TASK 2: Data Enrichment

```python
# average user population
population = []
for city_name in cities:
    populations = int(input(f"Enter the average population for {city_name}: "))
    population.append(populations)

print("\nCities:", cities)
print("Populations:", population)
```

# TASK 3: Data Analysis:

```python
# most populated city.
max_population  = max(population)
most_populated_city = cities[population.index(max_population)]
print("max_population:", max_population)
print("most_populated_city:",  most_populated_city)
```

```python
# least populated city.
min_population = min(population)
least_populated_city  = cities[population.index(min_population)]
print("min population:", min_population)
print("least_populated_city:", least_populated_city)
```

```python
# average population.
average_population = sum(population) / len(population)
print("average_population: ", average_population)
```

```python
# Add one more city and its population.
new_city = input("Enter the name of one more city to add: ")
new_population = int(input(f"Enter the average population for {new_city}: "))
cities.append(new_city)
population.append(new_population)
print(f"\nCities: {cities}")
print(f"Population: {population}")
```

```python
# Removing a single city
remove_city = input("\nEnter the name of the city to remove from the list: ")

if remove_city in cities:
    index_to_remove = cities.index(remove_city)
    cities.pop(index_to_remove)
    population.pop(index_to_remove)  # <-- use the correct variable name
    print(f"{remove_city} removed successfully.")
else:
    print(f"{remove_city} not found in the list.")
```

```python
#SUMMARY
print(f"After analysis, the most populated city is {most_populated_city} with an estimated number of {max_population:,} people.")
print(f"The least populated city is {least_populated_city} with an estimated number of {min_population:,} people.")
print(f"The estimated average population across all cities is {average_population:,.0f}.")
print(f"Updated city list: {cities}")

```
